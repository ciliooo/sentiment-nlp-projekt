```python
from pathlib import Path

DATA_DIR = Path("data")
FILES = list(DATA_DIR.glob("*.jsonl")) + list(DATA_DIR.glob("*.jsonl.gz"))

print("Gefundene Dateien:")
for f in FILES:
    print(" -", f.name)
```

```python
import pandas as pd

dfs = []
rows_per_file = 100_000   # 100k Zeilen pro Datei

for f in FILES:
    print("Lade:", f.name)
    df_part = pd.read_json(f, lines=True, nrows=rows_per_file)
    df_part["source_file"] = f.name  # merken
    dfs.append(df_part)

df = pd.concat(dfs, ignore_index=True)

print("\nGesamtanzahl Zeilen:", len(df))
df.head(3)
```

```python
# Nur Zeilen mit gültigem Rating 1–5
df_clean = df.copy()

# Sicherheit: Rating als Zahl erzwingen
df_clean["rating"] = pd.to_numeric(df_clean["rating"], errors="coerce")

# NaN raus und auf Bereich 1–5 begrenzen
df_clean = df_clean.dropna(subset=["rating", "text"])
df_clean = df_clean[(df_clean["rating"] >= 1) & (df_clean["rating"] <= 5)]

# Text zusammenbauen (Titel + Text, falls Titel fehlt, einfach "")
df_clean["title"] = df_clean["title"].fillna("")
df_clean["text"] = df_clean["text"].fillna("")

df_clean["full_text"] = (df_clean["title"].astype(str) + " " + df_clean["text"].astype(str)).str.strip()

# Leere Texte weg
df_clean = df_clean[df_clean["full_text"].str.len() > 0]

print("Bereinigte Zeilen:", len(df_clean))
print(df_clean[["rating", "title", "text", "source_file"]].head(5))
```
